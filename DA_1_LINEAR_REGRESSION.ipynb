```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import tree
from sklearn.model_selection import train_test_split
import seaborn as sns
import sklearn
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(df['cgpa'], df['placement_exam_marks'], hue=df['placed'])
plt.xlabel('cgpa')
plt.ylabel('placement_exam_marks')
plt.title('cgpa vs placement_exam_marks')
plt.show()
```

```python
X = df[['cgpa']]
X
```

```python
y = df[['placement_exam_marks']]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7,random_state=100)
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
print('Intercept of the Model:',lm.intercept_)
print('Co-efficient of the line:',lm.coef_)
c=lm.intercept_[0]
m=lm.coef_[0][0]
print("The linear Equation is: Y = {:.5}X + {:.5} ".format(c, m))
```

```python
y_predict=lm.predict(X_test)
y_predict
```

```python
plt.scatter (X_test, y_test, color = 'black')
plt.plot (X_test, lm.predict (X_test), color = 'blue')
plt.title('cgpa vs placement_exam_marks')
plt.xlabel('cgpa')
plt.ylabel('placement_exam_marks')
plt.show()
```

```python
from sklearn import metrics
print("Mean Absolute Error : ", metrics.mean_absolute_error(y_test, y_predict))
print('Mean Squared Error : ', metrics.mean_squared_error(y_test, y_predict))
print('Root Mean Squared Error : ', np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
print('R Square value : ', metrics.r2_score(y_test, y_predict))
```
